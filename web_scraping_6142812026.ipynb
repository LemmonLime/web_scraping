```python
import pandas as pd
import requests
import bs4
from bs4 import BeautifulSoup
from PIL import Image
import io
import re
```

```python
url='https://notebookspec.com/notebook/search?gamingMode=1'
r=requests.get(url)
r.text[:100]
```

```python
s = BeautifulSoup(r.text, 'lxml')
```

```python
print(s.prettify())
```

```python
nb = s.select('li.search-result-item') 
nb
```

```python
lst=[]
for e in nb:
    item_info=e.select_one('div.item-info').text.strip().replace('\n','')
    score=e.select_one('div.score').text.strip().replace('\n','')
    price=e.select_one('div.price').text.strip().replace('\n','')
    print(f'{item_info}  {score} {price}')
    lst.append([item_info , score, price])
lst[:1]
```

```python
df=pd.DataFrame(lst, columns=['Name' , 'Score', 'Price'])
df.head()
```

```python
da = df['Price'].str.extract(r'(\d+,\d+)').replace(r'[,]','',regex=True).astype(float)
da.columns=['base_price']
da
```

```python
db=df['Score']
db
```

```python
db=df['Score'].str.extract(r'\((\d{2}[.]\d{2})\)(.+)\((\d{2}[.]\d{2})\)(.+)')
# db.columns=['ประสิทธิภาพ','การเล่นเกม','งานกราฟิก','ความคุ้มค่า']
dq=db[0].astype(float)
dq
```

```python
de=db[2].astype(float)
de
```

```python
dw=db[1].str.extract(r'(\d{2}[.]\d{2})\)').astype(float)
dw
```

```python
dr=db[3].str.extract(r'(\d{2}[.]\d{2})\)').astype(float)
dr
```

```python
ds=pd.concat([dq,dw,de,dr], axis=1)
ds.columns=['ประสิทธิภาพ','การเล่นเกม','งานกราฟิก','ความคุ้มค่า']
ds
```

```python
d_clean=pd.concat([df, da, ds], axis=1)
d_clean
```
